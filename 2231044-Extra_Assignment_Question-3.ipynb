```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
age=int((input("Enter the age:")))
if (age>=18 and age<=25):
    print ("Adult")
elif (age>25 and age<=59):
    print ("Citizen")
elif (age >59):
    print ("Senior Citizen")
else:
    print("Children")
    
```
